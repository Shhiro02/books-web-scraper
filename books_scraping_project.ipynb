# 📚 Books Web Scraping Project

## Project Overview
This project demonstrates how to collect data from a website using **web scraping with Python**. The dataset is scraped from the practice website used for learning scraping: [https://books.toscrape.com](https://books.toscrape.com).

The goal of this project is to extract structured information about books such as:

- Book Title
- Price
- Rating
- Stock Availability

The extracted data is then stored in a dataset for further analysis.

The objectives of this project are:

## 📦 Libraries Used

### 1. requests
The `requests` library is used to send HTTP requests to a website and download the HTML content of a webpage.

Why it is necessary:
- Allows Python to communicate with websites
- Retrieves webpage content for parsing

### 2. BeautifulSoup
`BeautifulSoup` is used to parse HTML and navigate the webpage structure.

Why it is necessary:
- Helps locate specific HTML elements
- Extracts data using tags and CSS selectors

### 3. pandas
`pandas` is used to organize the extracted data into a structured dataset.

Why it is necessary:
- Converts raw data into DataFrames
- Allows exporting data into CSV format

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

## 🌐 Website Structure

Each page of the website contains multiple books. Each book is stored inside the following HTML container:

`article.product_pod`

Inside this container we can extract:

- Title → `h3 > a`
- Price → `p.price_color`
- Rating → `p.star-rating`
- Stock availability → `p.instock.availability`


## ⭐ Rating Mapping

The website stores ratings as text values such as:

- One
- Two
- Three
- Four
- Five

To make analysis easier, we convert them into numeric values.

```python
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}
```

## 🔎 Scraping Book Data

Steps performed:

1. Loop through all pages of the website
2. Send request to each page
3. Parse HTML content
4. Extract required book information
5. Store the data in a list


```python
data = []

for page in range(1, 51):
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")
    books = soup.select("article.product_pod")

    for book in books:
        title = book.h3.a['title']

        price = book.select_one("p.price_color").text.split("£")[1]

        rating_class = book.select_one(".star-rating")["class"][1]
        rating = rating_map[rating_class]

        stock = book.select_one(".instock.availability").text.strip()

        data.append([title, price, rating, stock])
    time.sleep(1)

print("Scraping Completed")
```

## 📊 Creating Dataset

After collecting all the book information, we convert the list into a **pandas DataFrame**.

This makes the data structured and easy to analyze.

```python
df = pd.DataFrame(data, columns=[
    "Title",
    "Price",
    "Rating",
    "Stock"
])

df.head()
```

## 💾 Saving the Dataset

The dataset is exported into a CSV file for further use in data analysis or machine learning projects.

```python
df.to_csv("books_dataset.csv", index=False)

print("Dataset saved successfully.")
```
